### Conjunto de Vetores

Uma coleo de vetores  chamada de **conjunto** e  representado por letras maiusculas em itlico. 

Exemplo:
$$
V = \begin{Bmatrix} v_1,v_2,...,v_n \end{Bmatrix}
$$

Um conjunto de vetores pode conter uma quantidade **finita ou infinita** de vetores. Quando contm infinitos vetores, ele pode formar um **subespao vetorial**, o que tem implicaes importantes na modelagem estatstica de dados.

Um conjunto de vetores tambm pode ser **vazio**, representado por $V=\begin{Bmatrix}\end{Bmatrix}$

### Combinao Linear Ponderada

Uma **combinao linear ponderada**  uma forma de integrar informaes de mltiplas variveis, onde algumas contribuem mais do que outras. s vezes, o termo *coeficiente*  utilizado em vez de *peso*.

A combinao linear ponderada consiste em multiplicaes escalar-vetor, seguidas por uma soma.

Presumimos que todos os vetores $v_i$ tem a mesma dimensionalidade e $\lambda_i$ podem ser qualquer nmero real.

Frmula geral:
$$
w = \lambda_1  v_1 + \lambda_2  v_2 + ... + \lambda_n  v_n
$$


**Aplicaes:**

- Em modelos estatsticos, as previses so obtidas por combinaes lineares ponderadas dos regressores e seus coeficientes (calculados, por exemplo, pelo mtodo dos mnimos quadrados).

- Em **PCA** (Anlise de Componentes Principais), os componentes principais so combinaes lineares dos dados, com pesos escolhidos para maximizar a varincia.

- Em **redes neurais artificiais**, a primeira operao  uma combinao linear ponderada das entradas, seguida de uma funo de ativao no linear. Os pesos so ajustados para minimizar uma funo de custo (geralmente o erro entre a previso e o valor real).

```python
# escalares
l1 = 1
l2 = 2
l3 = -3

# vetores
v1 = np.array([4,5,1])
v2 = np.array([-4,0,-4])
v3 = np.array([1,3,2])

# combinao linear ponderada
l1*v1 + l2*v2 + l3*v3
```

### Independncia linear

Um conjunto de vetores  **linearmente dependente** se ao menos um vetor puder ser expresso como uma combinao linear dos outros.

 **linearmente independente** se **nenhum** vetor puder ser expresso dessa forma.

$$
V = 
\begin{Bmatrix}
\begin{bmatrix}
1 \\ 3
\end{bmatrix},
\begin{bmatrix}
2 \\ 7
\end{bmatrix}
\end{Bmatrix},

\quad

S = 
\begin{Bmatrix}
\begin{bmatrix}
1 \\ 3
\end{bmatrix},
\begin{bmatrix}
2 \\ 6
\end{bmatrix}
\end{Bmatrix}

$$

- O conjunto V  **linearmente independente**.

- O conjunto S  **linearmente dependente**.

Para verificar a independncia linear, constri-se uma matriz com os vetores e calcula-se seu posto (rank). Se o rank for igual ao nmero de vetores, eles so linearmente independentes.

### A matemtica da independncia linear

*Dependncia Linear:*
$$
\bold{0} = \lambda_1  v_1 + \lambda_2  v_2 + ... + \lambda_n  v_n,
\quad
\lambda \in \mathbb{R}
$$

Essa equao diz que a dependncia linear pode ser definida por uma combinao linear ponderada dos vetores no conjunto que resulta em um vetor de zeros.

Logo:
$$
\lambda_1  v_1 = \lambda_2  v_2 + ... + \lambda_n  v_n,
\quad
\lambda \in \mathbb{R}
$$

- *Todo conjunto de vetores  dependente ou independente, no um vetor individual*

- *Pelo menos um $\lambda \neq 0$*

### Subespao e conjunto gerador

Um **subespao vetorial**  formado por **todas** as combinaes lineares possveis dos vetores de um conjunto.

O processo de formar essas combinaes  chamado de envoltria linear (ou espao gerado). O conjunto original  o conjunto gerador.

A dimenso de um subespao  o nmero mnimo de vetores linearmente independentes que o geram. Assim:

- Se os vetores forem linearmente independentes, a dimenso do subespao  igual ao nmero de vetores.

- Se forem dependentes, a dimenso ser menor que a quantidade de vetores.

**Definio formal:**
Um subespao vetorial  um subconjunto que  fechado sob adio e multiplicao escalar, e contm a origem do espao (o vetor nulo).

### Base

Uma **base**  um conjunto de vetores linearmente independentes que geram um subespao.

Ela funciona como um **sistema de coordenadas**: qualquer vetor no subespao pode ser descrito unicamente por uma combinao linear dos vetores da base.

O conjunto de base mais comum  o dos **eixos cartesianos**, como o plano XY.

No existe uma base melhor universalmente  a escolha da base depende do problema, variveis e contexto da anlise.

```python
import plotly.graph_objects as go

xlim = [-4,4]

# Dois vetores em um espao tridimensional
v1 = np.array([ 3,5,1 ])
v2 = np.array([ 0,2,2 ])


# Escalares aletorios no intervalo do eixo X
scalars = np.random.uniform(low=xlim[0],high=xlim[1],size=(100,2))



# Cria pontos aleatrios
points = np.zeros((100,3))
for i in range(len(scalars)):

  # Definir esse ponto como uma combinao aleatria ponderada dos dois vetores
  points[i,:] = v1*scalars[i,0] + v2*scalars[i,1]


# Desenhar os pontos no plano
fig = go.Figure( data=[go.Scatter3d(x=points[:,0], y=points[:,1], z=points[:,2], 
                                    mode='markers', marker=dict(size=6,color='black') )])

fig.update_layout(margin=dict(l=0,r=0,b=0,t=0))
```

**Resumindo:**

Um conjunto de vetores forma uma base de um subespao se:

- Ele abrange (gera) o subespao.

-  linearmente independente.

Isso garante que qualquer vetor no subespao pode ser descrito de forma nica usando essa base.
